```python
data_train=pd.read_csv('/content/Penelitian/Analisis_Sentimen_PPDB_2018_2022/Data/Labeled_data.csv', sep=';' ,encoding='utf-8')
from html import unescape
def clean_html(Komentar):
    Komentar = unescape(Komentar)
    return Komentar

data_train['Komentar'] = data_train['Komentar'].apply(lambda x: clean_html(x))
data_train.head(30) 
```

```python
data_train.info()
```

# Cleaning Text

```python
def remove_punct(Komentar):
  Komentar = Komentar.replace('\\t', " ").replace('\\n', " ").replace('\\u', " ").replace('\\', " ") # remove tab, new line, ans back slice
  Komentar = re.sub('(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)',' ', Komentar) # hapus mention
  Komentar = re.sub(r'^RT[\s]+', '', Komentar) # hapus retweet
  Komentar = re.sub(re.compile(r'(\d{3}[-\.\s]??\d{3}[-\.\s]??\d{4}|\(\d{3}\)\s*\d{3}[-\.\s]??\d{4}|\d{3}[-\.\s]??\d{4})'),' ', Komentar) # hapus no.telepon
  Komentar = re.sub(r'((www\.[^\s]+)|(https?://[^\s]+)|(http?://[^\s]+))',' ', Komentar) #hapus link, URL
  Komentar = re.sub(r' +', ' ', Komentar)
  Komentar = re.sub(r"\b[a-zA-Z]\b", "", Komentar) #menghapus data tunggal
    
  return Komentar

pd.set_option("display.max_colwidth", -1)
data_train['KOMENTAR'] = data_train['Komentar'].apply(lambda x: remove_punct(x))
data_train.head(150)
```

# Tokenisasi

```python
def tokenization(Komentar):
    Komentar = re.split('\W+', Komentar)
    return Komentar

data_train['TOKENIZATION'] = data_train['KOMENTAR'].apply(lambda x: tokenization(x.lower()))
data_train.head()
```

# Stop Removal

```python
stopword = nltk.corpus.stopwords.words('indonesian')

def remove_stopwords(Komentar):
    Komentar = [word for word in Komentar if word not in stopword]
    return Komentar

data_train['STOP_REMOVAL'] = data_train['TOKENIZATION'].apply(lambda x: remove_stopwords(x))
data_train.head()
```

# Case Folding

```python
#case folding
data_train['KOMENTAR'] = data_train['Komentar'].str.lower()
data_train.head()
```

# Mengubah ke bentuk teks

```python
stop_removal = data_train[['STOP_REMOVAL']]

def fit_stopwords(text):
    text = np.array(text)
    text = ' '.join(text)
    return text

data_train['STOP_REMOVAL'] = data_train['STOP_REMOVAL'].apply(lambda x: fit_stopwords(x))
data_train.head(150)
```
