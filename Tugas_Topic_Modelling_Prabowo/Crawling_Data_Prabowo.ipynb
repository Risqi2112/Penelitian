```python
from datetime import datetime, timedelta
from duckduckgo_search import ddg
import pandas as pd
```

```python
def duckweb(keywords='', N=30, lang='Id-Id', time=''):
    return ddg(keywords, region=lang, safesearch='On', time=time, max_results=N)
```

```python
start_date = datetime(2023, 3, 1)  # Tanggal mulai
end_date = datetime(2023, 3, 11)  # Tanggal akhir
```

```python
delta = timedelta(days=1)  # Jangka waktu
```

```python
df_list = []
while start_date < end_date:
    next_date = start_date + delta  # Tanggal berikutnya
    time_range = f"{start_date.strftime('%Y-%m-%d')}..{next_date.strftime('%Y-%m-%d')}"  # Rentang waktu untuk pencarian
    query = '"Prabowo Subianto" "Ketua Umum Gerindra" "Capres Gerindra" "Menteri Pertahanan" "Prabowo Presiden 2024" site:twitter.com'
    results = duckweb(keywords=query, time=time_range,lang ='Id-Id')

    df = pd.DataFrame(results)
    
    df['date'] = start_date.strftime('%Y-%m-%d')
    df_list.append(df)

    start_date = next_date  # Setel tanggal berikutnya sebagai tanggal mulai

df_1 = pd.concat(df_list)
```
