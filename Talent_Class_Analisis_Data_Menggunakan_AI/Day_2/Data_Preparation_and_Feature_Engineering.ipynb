## Fill with mean

```python
# Read file
df3 = pd.read_csv('https://raw.githubusercontent.com/Risqi2112/Penelitian/main/Talent_Class_Analisis_Data_Menggunakan_AI/Day_2/diabetes_null.csv', na_values=['#NAME?'])
df3.head()
```

```python
# Check empty values
df3.isna().sum()
```

```python
df3['Glucose'].fillna(df3['Glucose'].mean()).reset_index()
```

```python
df3['BloodPressure'].fillna(df3['BloodPressure'].mean()).reset_index()
```

```python
df3['SkinThickness'].fillna(df3['SkinThickness'].mean()).reset_index()
```

```python
df3['Insulin'].fillna(df3['Insulin'].mean()).reset_index()
```

```python
df3['BMI'].fillna(df3['BMI'].mean()).reset_index()
```
